```python
from parseHtrace import Parse, hash_tree, clean_nodes
import pandas as pd
pd.set_option('display.max_rows', None)
```

### wordcount

```python
try:
    w_tree_1 = pd.read_pickle('pickle/wordcount_tiny_datanode1.pkl')
except:
    wordcount = Parse('wordcount_tiny_datanode1.out')
    wordcount.build_tree()
    trees = [node for node in wordcount.trees]
    hashtree = hash_tree(trees)
    se_tree = pd.Series(hashtree)
    w_tree_1 = se_tree.value_counts().to_frame()
    w_tree_1.to_pickle('pickle/wordcount_tiny_datanode1.pkl')
try:
    w_tree_2 = pd.read_pickle('pickle/wordcount_tiny_datanode2.pkl')
except:
    wordcount = Parse('wordcount_tiny_datanode2.out')
    wordcount.build_tree()
    trees = [node for node in wordcount.trees]
    hashtree = hash_tree(trees)
    se_tree = pd.Series(hashtree)
    w_tree_2 = se_tree.value_counts().to_frame()
    w_tree_2.to_pickle('pickle/wordcount_tiny_datanode2.pkl')
```
